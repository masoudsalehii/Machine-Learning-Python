```python
# There are 23 houses without address which is only an inconsiderable number in 3479 
# Regrarding the importance of Address in price defination for a house we could remove these houses from our dataset.
df = df.dropna()
```

```python
# Now we can use our data; because the rows with missing addresses and irrational prices has been removed.
# Let's take 80% of data for train , and 20% of data for test
msk = np.random.rand(len(df))< 0.8
train = df[msk]
test = df[~msk]
y_train = train[['Price']]
y_test = test[['Price']]
# I think some features are highly effective in defining a house price such as address, area;
#So I only use them to train the model
x_train = train[['Area','Parking','Warehouse', 'Elevator', 'Address']]
x_test = test[['Area','Parking','Warehouse', 'Elevator', 'Address']]
#x_test['Address'] = pd.factorize(x_test['Address'])[4]
x_test.head()
```

```python
#To train our model, first we have to normalize our data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
y_test = scaler.fit_transform(y_test)
y_train = scaler.fit_transform(y_train)
# We must map the addresses to numbers in order to train the model
x_train['Address'] = pd.factorize(x_train['Address'])[0]
x_test['Address'] = pd.factorize(x_test['Address'])[0]
x_test[["Area","Address"]] = scaler.fit_transform(x_test[["Area","Address"]])
x_train[["Area","Address"]] = scaler.fit_transform(x_train[["Area","Address"]])
```

```python
# Normalizing our data and dividing them to the train and test set, it is time to train the model.
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
y_hat = model.predict(x_test)
```

```python
# Training the model, let's evaluate it:
from sklearn.metrics import r2_score
result = r2_score(y_test, y_hat)
print(result)
```
